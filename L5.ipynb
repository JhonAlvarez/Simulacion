# POISSON

```python
def POISSON(n,l,k):
    x=range(0,k+1)
    px=[]
    for i in x:
        px.append(((np.e**(-l))*(l**i))/math.factorial(i))
    ac = []
    ac.append(px[0])
    for i in range(1,len(px)):
        ac.append(px[i]+ac[i-1])
    plt.plot(px,linewidth=0, marker='o')
    plt.show()
    al = congruencial(n)
    rta=[]
    for i in al:
        for j in range(0,len(ac)):
            if i<=ac[j]:
                rta.append(x[j])
                break
                
    hist1 = plt.hist(rta, bins=k)
    plt.xlim(0.0, 20.0)
    plt.show()
    prob = []
    for i in hist1[0]:
        prob.append(float(i)/n)
    plt.plot(prob,linewidth=0, marker='o')
    plt.show()
    
POISSON(100000,9,17)    
```
