# Generador de variables aleatorias continuas y comparación de los resultados con las funciones de densidad

Tomando como distribuciones a trabajar la distribución logistica, weibull y uniforme continua.

## Para la distribución logistica, su función de densidad e  inversa:

$$ f(x; \mu,s) = \frac{e^{-(x-\mu)/s}} {s\left(1+e^{-(x-\mu)/s}\right)^2} \ $$






```python
import matplotlib.pyplot as plt
import math
import numpy as np
%matplotlib inline 

def gen_num(n):
    x0 = 123456
    a=22695477
    b=1
    m=2**32
    num = []
    for i in range(0,n):
        xi = ((a*x0)+(b))%m
        num.append(float(xi)/float(m))
        x0=xi
    return num


s=2
m=5

x1 = np.arange(-20,20,0.1)
logistica = []
for i in x1:
    euler = np.e**(-(i-m)/s)
    logistica.append(euler/(s*(1+euler)**2))
plt.plot(x1,logistica)
#plt.hist(inv, normed=True)
plt.show()
```

función inversa:

$$F^{-1}(p; \mu,s) = \mu + s\,\ln\left(\frac{p}{1-p}\right)$$

```python
x = gen_num(1000)
inv = []
for i in x:
    inv.append(float(m+(s*np.log(float(i)/float((1-i))))))
plt.hist(inv)
plt.show()
```

Ahora comparamos los resultados

```python
plt.plot(x1,logistica)
plt.hist(inv, normed=True)
plt.show()
```

## Para la distribución weibull, su función de densidad e inversa:

$$ f(x;\lambda,k) =  \begin{cases}
\frac{k}{\lambda}\left(\frac{x}{\lambda}\right)^{k-1}e^{-(x/\lambda)^{k}} & x\geq0\\
0 & x<0\end{cases} $$



```python
l=1
k=5
q=2
x1 = np.arange(0,q,0.1)
weibull = []
for i in x1:
    euler = np.e**(-(float(i)/l)**k)
    kl = (float(k)/float(l))*(float(i)/l)**(k-1)
    weibull.append(kl*euler)
plt.plot(x1,weibull)
plt.show()
```

función inversa:

$$ F^{-1}(p;k,\lambda) = \lambda {(-\ln(1-p))}^{1/k} $$

```python
x = gen_num(1000)
inv = []
for i in x:
    inv.append(float(l)*(-np.log(1.0-i))**(1.0/k))
plt.hist(inv)
plt.show()
```

Ahora comparamos los resultados

```python
plt.plot(x1,weibull)
plt.hist(inv, normed=True)
plt.show()
```

## Para la distribución uniforme continua, su función de densidad e inversa:

$$f(x)=\left\{\begin{matrix}
  \frac{1}{b - a} & \ \ \ \mathrm{para}\ a \le x \le b, \\  \\
  0 & \mathrm{para}\ x<a\ \mathrm{o}\ x>b, \end{matrix}\right.$$

```python
a=0
b=1
x1 = np.arange(-0.5,1.5,0.01)
uniforme = []
for i in x1:
    if a<= i and b>=i:
        uniforme.append(1.0/(b-a))
    else:
        uniforme.append(0)
plt.plot(x1,uniforme)
plt.ylim(0.0, 1.1)
plt.show()
```

función inversa:

$$F^{-1}(p) = a + p (b - a) \,\,\text{ for } 0<p<1$$

```python

x = gen_num(1000)
inv = []
for i in x:
    inv.append(a+i*(b-a))
plt.hist(inv)
plt.show()
```

Ahora comparamos los resultados

```python
plt.ylim(0.0, 1.5)
plt.plot(x1,uniforme)
plt.hist(inv, normed=True)
plt.show()
```
